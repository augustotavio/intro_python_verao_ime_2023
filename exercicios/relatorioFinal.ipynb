```python
"""     Oii, vou deixar aqui um contexto da minha tentativa de análise de dados rs. Queria fazer algo relacionado a meio ambiente
    e  desmatamento, e achei dados relacionado as Unidade de Conservação Federal. Econtrei três base de dados sobre três dados 
    diferentes e tentei relaciona-los: Desmatamento em UCs, Incêncios em UCs, e, uma base de dados mais completa sobre com infos 
    sobre essas UCs, e as tabelas se relacionavam através de um código de cada unidade de coversação. """

import pandas as pd

df_desm = pd.read_csv('../dados/desmatamento_ucs_atualizado_2021.csv', sep=';')

df_incen = pd.read_csv('../dados/incendios_em_unidades_de_conservacao_federal.csv', sep='\t')

df_lim_uc = pd.read_csv('../dados/Limite_UC_federais_jun_2021_CSV.csv', sep=';')

```

```python
# Na tabela sobre incêndios, veio mais uma coluna sem nova e sem dados mas sempre criava ela, ai aqui exclui 
# Infelizmente meio ambiente é apenas um assunto que tenho entresse mas não domino questões técnicas
# Nesse df, eu confesso que tentei pesquisar mas não consegui entender sobre os que se tratam as colunas AAF e o Ano em si, se é ...
# o ultimo ano registrado incêncio

df_incen2 = df_incen.copy()

df_incen2.drop(["Unnamed: 4"], axis=1, inplace=True)

df_incen2
```

```python
df_incen2.sort_values(by=['Cód. CNUC'])
```

```python
# Aqui criei um novo df que continha dados sobre as UCs e sobre os incêncios, relacionados através do código de cada UC

df_lim_incen = pd.merge(df_lim_uc,df_incen2, how='inner', on="Cód. CNUC")

df_lim_incen2 = df_lim_incen.copy()
df_lim_incen2.sort_values(by=["Cód. CNUC"])
```

```python
#aqui crei uma nova tabela que juntasse também os dados referentes a desmatamento

df_lim_incen_desma = pd.merge(df_lim_incen2,df_desm, how='inner', on="Cód. CNUC")

df_lim_incen_desma2 = df_lim_incen_desma.copy()

df_lim_incen_desma2
```

```python
# Exclui dados duplicados como nome das UC e o bioma referente, e exclui alguns que não iria utilizar, busquei focar..
# nos últimos 5 anos

df_lim_incen_desma2.drop(["Coord. Regional", "Ato legal", "Grupo", "Fuso/abrangencia", "UC", "uc", "biocgfun", "2016",	"2015",	"2014",	"2013",	"2012",	"2011",	"2010",	"2009", "2008",	"2007",	"2006",	"2005",	"2004",	"2003",	"2002",	"2001",	"2000",	"1997"], axis=1, inplace=True)
```

```python
df_lim_incen_desma2.sort_values(by=["Cód. CNUC"])
```

```python
# Aqui tentei criar uma coluna que somasse o desmatamento total do últimos 5 anos, e percebi que tinha um problema de tipo de dado

df_lim_incen_desma2['Desmatamento últimos 5 anos'] = df_lim_incen_desma2["2021"] +  df_lim_incen_desma2["2020"] + df_lim_incen_desma2["2019"] + df_lim_incen_desma2["2018"] +  df_lim_incen_desma2["2017"]

df_lim_incen_desma2
```

```python
# Aqui tentei alterar o tipo dos dados de colunas específicas para conseguir somar mas não consegui, pesquisei mas não consegui

df_lim_incen_desma2[[" Área (hect) ", " Perimetro (m) ", "AAF", "ANO", "2017", "2018", "2019", "2020", "2021"]] = df_lim_incen_desma2[[" Área (hect) ", " Perimetro (m) ", "AAF", "ANO", "2017", "2018", "2019", "2020", "2021"]].apply(pd.to_numeric)

df_lim_incen_desma2.dtypes
```
